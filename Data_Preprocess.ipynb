#Normalised Data


```
df = pd.read_csv('/mydrive/ManojSirIntern/data_final.csv')
df.drop(['Unnamed: 0'],axis =1,inplace=True)
Y = df['OutPut']
df.drop(['OutPut'],axis=1,inplace=True)
```

```
mt = np.array(df)
mt.shape
```

```
for i in range(len(mt)):
  mx = mt[i].max()
  mn = mt[i].min()
  if( (mx-mn)==0):
    continue
  for j in range(len(mt[i])):
    mt[i][j] = (mt[i][j]-mn)/(mx-mn)
```

```
data = pd.DataFrame(mt)
data
```

```
data['OutPut'] = Y
```
